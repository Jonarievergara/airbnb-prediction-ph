```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet

from sklearn.ensemble import RandomForestRegressor

from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import mean_absolute_error
from math import sqrt
from sklearn.metrics import r2_score
```

```python
pd.set_option('display.max_columns', 500)
```

```python
df=pd.read_csv("finaldata.csv")
```

```python
df.head()
```

```python
X = df.drop(columns=['url','rating','type_airbnb',\
                    'price_per_night',\
                    'province', 'is_shared_room', 'is_private_room', 'num_nearby_tourist_spots_3km',\
                    'num_nearby_supermarkets_3km', 'num_nearby_restaurants_3km','entire_home',\
                    'latitude', 'longitude',\
                    'population','area','covid_total','covid_active',\
                    'type_airbnb_label', 'province_label', 'foreign_travelers',\
                    'rating_cleanliness','rating_accuracy','rating_communication','rating_location','rating_checkin','rating_value'])
```

```python
X.head()
```

```python
X.shape
```

```python
scaler = StandardScaler()
scaled_features=scaler.fit_transform(X)
```

```python
X = pd.DataFrame(scaled_features, index=X.index, columns=X.columns)
```

```python
X.head()
```

```python
Y=df['price_per_night']
Y
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, Y,test_size = 0.2, random_state = 100)
```

# Linear Regression model

```python
linear_model = LinearRegression(normalize=False)
linear_model.fit(train_X,train_y)
pred_linear_train= linear_model.predict(train_X)
pred_linear_test= linear_model.predict(test_X)
```

```python
rmse = (np.sqrt(MSE(train_y, pred_linear_train)))
mse = (MSE(train_y, pred_linear_train))
r2 = r2_score(train_y, pred_linear_train)
print('Training performance')
print('RMSE: {:.2f}'.format(rmse))
print('MSE: {:.2f}'.format(mse))
print('R2: {:.2f}'.format(r2))
print('')
rmse = (np.sqrt(MSE(test_y, pred_linear_test)))
mse = (MSE(test_y, pred_linear_test))
r2 = r2_score(test_y, pred_linear_test)
print('Testing performance')
print('RMSE: {:.2f}'.format(rmse))
print('MSE: {:.2f}'.format(mse))
print('R2: {:.2f}'.format(r2))
```

```python
linear_cv=cross_val_score(linear_model, train_X, train_y)
print("Cross val (5 folds):", np.mean(linear_cv))
```

# Ridge Regression model

```python
ridge_model = Ridge(alpha = 0.01, normalize = False)
ridge_model.fit(train_X,train_y) 
pred_ridge_train= ridge_model.predict(train_X)
pred_ridge_test= ridge_model.predict(test_X)
```

```python
rmse = (np.sqrt(MSE(train_y, pred_ridge_train)))
mse = (MSE(train_y, pred_ridge_train))
r2 = r2_score(train_y, pred_ridge_train)
print('Training performance')
print('RMSE: {:.2f}'.format(rmse))
print('MSE: {:.2f}'.format(mse))
print('R2: {:.2f}'.format(r2))
print('')
rmse = (np.sqrt(MSE(test_y, pred_ridge_test)))
mse = (MSE(test_y, pred_ridge_test))
r2 = r2_score(test_y, pred_ridge_test)
print('Testing performance')
print('RMSE: {:.2f}'.format(rmse))
print('MSE: {:.2f}'.format(mse))
print('R2: {:.2f}'.format(r2))
```

```python
ridge_model_cv=cross_val_score(ridge_model, train_X, train_y)
print("Cross val (5 folds):", np.mean(ridge_model_cv))
```

```python
lasso_model = Lasso(alpha = 0.001, normalize =False)
lasso_model.fit(train_X,train_y) 
pred_lasso_train = lasso_model.predict(train_X)
pred_lasso_test= lasso_model.predict(test_X)
```

```python
lasso_model_cv=cross_val_score(lasso_model, train_X, train_y)
print("Cross val (5 folds):", np.mean(lasso_model_cv))
```
